### Cleaning Data

#### Menangani Missing Value pada orders_df

```python
orders_df[orders_df.order_approved_at.isna()]
```

```python
orders_df.order_approved_at.value_counts()
```

```python
orders_df["order_approved_at"].fillna(orders_df["order_approved_at"].mode()[0], inplace=True)
```

```python
orders_df.isna().sum()
```

```python
orders_df[orders_df.order_delivered_carrier_date.isna()]
```

```python
orders_df["order_delivered_carrier_date"].fillna(orders_df["order_delivered_carrier_date"].mode()[0], inplace=True)
```

```python
orders_df.isna().sum()
```

```python
orders_df[orders_df.order_delivered_customer_date.isna()]
```

```python
orders_df["order_delivered_customer_date"].fillna(orders_df["order_delivered_customer_date"].mode()[0], inplace=True)
```

```python
orders_df.isna().sum()
```

#### Mengganti tipe data pada orders_df

```python
columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 
           'order_delivered_customer_date', 'order_estimated_delivery_date']

for col in columns:
    orders_df[col] = pd.to_datetime(orders_df[col])
```

```python
orders_df.info()
```

#### Mengganti tipe data order_items_df

```python
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])
```

```python
order_items_df.info()
```

#### Menangani missing value pada products_df

```python
all_columns = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
for column in all_columns:
    products_df[column].fillna(products_df[column].median(), inplace=True)
```

```python
products_df.isna().sum()
```

```python
new_columns = ['product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty']

for column in new_columns:
    most_common_value = products_df[column].mode()[0]
    products_df[column].fillna(most_common_value, inplace=True)
```

```python
products_df.isna().sum()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Harga Penjualan Produk dan Pendapatan di E-Commerce dalam beberapa bulan belakangan?

```python
all_df["order_purchase_timestamp"] = pd.to_datetime(all_df["order_purchase_timestamp"])
data_2017 = all_df[all_df['order_purchase_timestamp'].dt.year == 2017]
order_recently_df = data_2017.resample(rule="M", on="order_purchase_timestamp").agg({
    "order_id": "nunique",
    "price": "sum"
})
```

```python
order_recently_df.index = order_recently_df.index.strftime('%Y-%m')
order_recently_df = order_recently_df.reset_index()
order_recently_df.rename(columns={
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)
order_recently_df.head()
```

```python
data_2017 = all_df[all_df['order_purchase_timestamp'].dt.year == 2017]
order_recently_df = data_2017.resample(rule="M", on="order_purchase_timestamp").agg({
    "order_id": "nunique",
    "price": "sum"
})
order_recently_df.index = order_recently_df.index.strftime('%B')
 
order_recently_df = order_recently_df.reset_index()
order_recently_df.rename(columns={
    "order_id": "count_order",
    "price": "revenue"
}, inplace=True)
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(order_recently_df["order_purchase_timestamp"], order_recently_df["count_order"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Orders per Month (2017)", loc="center", fontsize=20), plt.xticks(rotation=45) 
```

```python
plt.figure(figsize=(10, 5)) 
plt.bar(order_recently_df["order_purchase_timestamp"], order_recently_df["count_order"], color="#72BCD4") 
plt.title("Number of Orders per Month (2017)", loc="center", fontsize=20) 
plt.xticks(rotation=45)
plt.xlabel("Month", fontsize=14)
plt.ylabel("Number of Orders", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    order_recently_df["order_purchase_timestamp"],
    order_recently_df["revenue"],
    marker='o', 
    linewidth=2,
    color="#72BCD4"
)
plt.title("Total Revenue per Month (2017)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(rotation=45)
plt.show()
```
